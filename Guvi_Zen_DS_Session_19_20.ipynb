1.   **Box Plot (Box-and-Whisker Plot)**:

Description: A box plot displays the distribution of a dataset by showing the median, quartiles, and potential outliers.

When to Use: Use box plots to visualize the spread and skewness of data and to identify outliers or compare distributions between multiple groups.

Data: Typically used for numerical data with outliers.

2.  **Pie Chart**

Description: A circular chart divided into sectors, where each sector represents a proportion of a whole.

When to Use: Use pie charts to show the composition of a whole, such as percentages of a total.

Data: Categorical data with relative proportions.

3.   **Donut Plot**:

Description: Similar to a pie chart but with a hole in the center. It provides a cleaner look while conveying the same information.

When to Use: Use donut plots when you want to show the composition of a whole, similar to pie charts.

Data: Categorical data with relative proportions.

4.   **Stacked Bar Plot**

Description: A bar chart where each bar is divided into segments, and each segment represents a subcategory or component.

When to Use: Use stacked bar plots to show the composition of a whole for multiple categories and compare the contribution of each subcategory.

Data: Categorical data with subcategories.

5.   **Relative Stacked Bar Plot**:

Description: A variation of the stacked bar plot, where each bar's height represents the total, and segments represent the relative contributions of subcategories.

When to Use: Use relative stacked bar plots when you want to emphasize the proportions within each category.

Data: Categorical data with subcategories.

6.  ** Stacked Area Plot**:

Description: Similar to a line plot but with areas under the lines filled to represent data proportions over time.

When to Use: Use stacked area plots to visualize how different categories or components contribute to a changing total over time.

Data: Time-series data with categorical or numerical components.

7.   **Scatter Plots**

Description: Scatter plots represent individual data points with two variables on the x and y axes, showing their relationships.

When to Use: Use scatter plots to explore correlations, clusters, or outliers between two variables.

Data: Two numerical variables.

8.   **Bar Plot**:

Description: A chart that represents data with rectangular bars, where the length or height of each bar corresponds to the value.

When to Use: Use bar plots to compare data across different categories or to display frequencies.

Data: Categorical or numerical data with comparisons.

9.  **Continuous vs Continuous Plot:**

Description: A plot that displays a relationship between two continuous variables, such as a scatter plot or a line plot.

When to Use: Use continuous vs continuous plots to analyze the relationship, trends, and correlations between two continuous variables.

Data: Two numerical variables.

10.   **Line Plot:**

Description: A chart that connects data points with lines, often used to represent trends over time.

When to Use: Use line plots to visualize data over time or to show trends and patterns.

Data: Time-series data or numerical data over continuous intervals.


The choice of the right data visualization plot depends on your specific data, your objectives, and the insights you want to convey. Each plot has its strengths and weaknesses, so it's important to choose the one that best suits your analysis and communication goals.


A histogram is a graphical representation of the distribution of a dataset. It divides the data into bins or intervals and displays the frequency (or count) of data points in each bin. Histograms are useful for visualizing the shape, central tendency, and spread of a dataset.

```python
# Histogram

import matplotlib.pyplot as plt
import numpy as np

# Generate example data (randomly generated data for demonstration)
data = np.random.randn(100000)  # Create a dataset of 1000 random values

# Create a histogram
plt.hist(data, bins=20, color='#2fd611', edgecolor='#fc0331')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Display the histogram
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a useful tool for visualizing the distribution of a dataset and showing key summary statistics.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate example data (randomly generated data for demonstration)
data = [np.random.randn(1000) for _ in range(5)]

print(len(data[0]))
data[0] = data[0]+2 # shifting the mean by 2
data[3] = data[3]+2.5 # shifting the mean by 2.5

# Create a box plot
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5'])
plt.title('Box Plot of Data')
plt.xlabel('Groups')
plt.ylabel('Values')

# Display the box plot
plt.show()
```

A pie chart is a circular statistical graphic that is divided into slices to represent the proportion of different categories within a dataset

```python
import matplotlib.pyplot as plt

# Example data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 15, 55, 20]  # Sizes or proportions for each category
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']  # Color for each category

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0)

# Add a title
plt.title('Distribution of Categories')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is circular.
plt.show()
```

A donut chart is a variation of a pie chart with a hole in the center.

```python
import matplotlib.pyplot as plt

# Example data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 15, 45, 10]  # Sizes or proportions for each category
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']  # Colors for each category

# Create the outer pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, radius=1)

# Draw a white circle in the center to create the donut hole
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

# Add a title
plt.title('Donut Chart Example')

# Display the donut chart
plt.axis('equal')
plt.show()
```

A stacked bar plot is a data visualization that displays multiple datasets as horizontal or vertical bars, with each bar divided into segments or categories.

```python
import matplotlib.pyplot as plt

# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data1 = [10, 15, 5, 8]
data2 = [5, 10, 15, 7]
data3 = [12, 6, 8, 10]
data4 = [10,12,8,9]

# Create a stacked bar plot
plt.bar(categories, data1, label='Dataset 1', color='tab:blue')
plt.bar(categories, data2, label='Dataset 2', color='tab:orange', bottom=data1)
plt.bar(categories, data3, label='Dataset 3', color='tab:green', bottom=[i + j for i, j in zip(data1, data2)])
plt.bar(categories, data4, label='Dataset 4', color='tab:red', bottom=[i + j + k for i, j, k in zip(data1, data2, data3)])

# Add a legend
plt.legend()

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot Example')

# Display the stacked bar plot
plt.show()
```

A relative stacked bar plot, often referred to as a stacked 100% bar plot, displays the relative proportions of different datasets as percentages of the whole. This type of plot can be useful for comparing the composition of datasets, where each bar represents 100% of the total, and the segments within the bars show the relative proportions of different categories.

```python
import matplotlib.pyplot as plt

# Example data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data1 = [10, 15, 5, 8]
data2 = [5, 10, 15, 7]
data3 = [12, 6, 8, 10]

# Calculate the total for each category
totals = [i + j + k for i, j, k in zip(data1, data2, data3)]

# Calculate the percentages for each dataset
data1_percent = [i / j * 100 for i, j in zip(data1, totals)]
data2_percent = [i / j * 100 for i, j in zip(data2, totals)]
data3_percent = [i / j * 100 for i, j in zip(data3, totals)]

# Create a relative stacked bar plot
plt.bar(categories, data1_percent, label='Dataset 1', color='tab:blue')
plt.bar(categories, data2_percent, label='Dataset 2', color='tab:orange', bottom=data1_percent)
plt.bar(categories, data3_percent, label='Dataset 3', color='tab:green', bottom=[i + j for i, j in zip(data1_percent, data2_percent)])

# Add a legend
plt.legend()

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Percentage (%)')
plt.title('Relative Stacked Bar Plot Example')

# Display the relative stacked bar plot
plt.show()
```

A stacked area plot, also known as a stacked area chart, is a data visualization that shows the composition and trends of multiple datasets over a continuous or discrete domain. It's useful for displaying how different categories or variables contribute to a whole over time or another dimension.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.arange(0, 11)  # x-axis values
y1 = np.array([3, 6, 9, 12, 15, 12, 9, 6, 3, 6, 9])  # Dataset 1
y2 = np.array([1, 2, 3, 4, 5, 4, 3, 2, 1, 2, 3])  # Dataset 2
y3 = np.array([2, 4, 6, 8, 10, 8, 6, 4, 2, 4, 6])  # Dataset 3

# Create a stacked area plot
plt.stackplot(x, y1, y2, y3, labels=['Dataset 1', 'Dataset 2', 'Dataset 3'], baseline='zero',alpha=0.5)

# Add labels, title, and legend
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Stacked Area Plot Example')
plt.legend(loc='upper right')

# Display the stacked area plot
plt.show()
```

Scatter plots are used to visualize the relationship between two continuous variables. They are particularly useful for identifying patterns, trends, and relationships within data

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.random.rand(50)  # Generate 50 random values for the x-axis
y = 2 * x + 1 + np.random.randn(50)  # Generate corresponding values for the y-axis with noise

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='blue', marker='+', label='Data Points')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Scatter Plot Example')

# Add a legend
plt.legend()

# Display the scatter plot
plt.grid(True)
plt.show()
```

A line plot is a type of data visualization that displays data points as individual marks and connects them with lines. Line plots are typically used to represent trends, changes over time, or to show the relationship between two continuous variables.

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)  # Generate 100 values for the x-axis
y = np.sin(x)  # Generate corresponding values for the y-axis (sinusoidal function)
#y = 2+x+np.random.randn(100)
# Create a line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='green', linestyle='-', marker='o', markersize=5, label='Sine Curve')

# Add labels and title
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plot Example')

# Add a legend
plt.legend()

# Display the line plot
plt.grid(True)
plt.show()
```

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file 'district.csv' into a Pandas DataFrame
data = pd.read_csv('district.csv')

# Extract the first 30 rows of specific columns from the DataFrame:
# - 're' represents a column from the 6th index (1-based) containing recovery data
# - 'de' represents a column from the 5th index (1-based) containing deceased data
# - 'co' represents a column from the 4th index (1-based) containing confirmed cases data
# - 'x' represents a column from the 1st index (0-based) containing district names
re = data.iloc[:30, 5].values  # Extract values from the 31st row (0-based index)
de = data.iloc[:30, 4].values  # Extract values from the 31st row
co = data.iloc[:30, 3].values
x = list(data.iloc[:30, 0])

# Create a figure and axis for the plot
plt.figure(figsize=(25, 10))
ax = plt.axes()

# Customize the plot background and grid
ax.set_facecolor('black')
ax.grid(linewidth=0.4, color='#8f8f8f')

# Customize X and Y ticks appearance
plt.xticks(rotation='vertical', size='20', color='green')  # X-axis ticks
plt.yticks(size='20', color='red')  # Y-axis ticks

# Customize axis labels
ax.set_xlabel('\nDistrict', size=25, color='#4bb4f2')
ax.set_ylabel('No. of cases\n', size=25, color='#4bb4f2')

# Customize tick parameters
plt.tick_params(size=20, color='white')

# Set the title of the plot
ax.set_title('Maharashtra District wise breakdown\n', size=50, color='#28a9ff')

# Create a bar plot for each set of data
plt.bar(x, co, label='re')
plt.bar(x, re, label='re', color='green')
plt.bar(x, de, label='re', color='red')

# Annotate the bars with the respective data values
for i, j in zip(x, co):
    ax.annotate(str(int(j)),
                xy=(i, j + 3),
                color='white',
                size='15')

# Add a legend to the plot
plt.legend(['Confirmed', 'Recovered', 'Deceased'], fontsize=20)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Generate example data
data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value1': np.random.randn(100),
    'Value2': np.random.randint(1, 100, size=100),
    'Value3': np.random.uniform(0, 1, size=100),
}

df = pd.DataFrame(data)

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Value1', bins=20, kde=True)
plt.title('Histogram')
plt.show()

# Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Value2')
plt.title('Box Plot')
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Value1', y='Value2', hue='Category')
plt.title('Scatter Plot')
plt.show()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()

# Donut Chart
plt.figure(figsize=(8, 8))
plt.pie(df['Category'].value_counts(), labels=df['Category'].value_counts().index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Donut Chart')
plt.show()

# Stacked Bar Plot
stacked_data = pd.crosstab(index=df['Category'], columns='count')
stacked_data.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Stacked Bar Plot')
plt.show()

# Relative Stacked Bar Plot
relative_stacked_data = stacked_data.div(stacked_data.sum(axis=1), axis=0)
relative_stacked_data.plot(kind='bar', stacked=True, figsize=(8, 5))
plt.title('Relative Stacked Bar Plot')
plt.show()

# Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Category', y='Value2')
plt.title('Bar Plot')
plt.show()

# Line Plot
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Value2', y='Value1', hue='Category')
plt.title('Line Plot')
plt.show()

# Continuous Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Value1', hue='Category', fill=True)
plt.title('Continuous Plot')
plt.show()
```
